### Explanatory Data Analysis For Machine Learning in Egypt

```python
#Libraries imported:

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datasets import load_dataset
import ast

#Loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Clean Up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Filtering for Egypt Machine Learning roles

```python
df_ML_EGY = df[(df['job_country'] == 'Egypt') & (df['job_title_short'] == 'Machine Learning Engineer')]
```

### Locations to explore:

```python
df_plot = df_ML_EGY['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.despine()
plt.title('Counts of job locations for Machine Learning roles in Egypt')
plt.ylabel('')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from home',
    'job_no_degree_mention': 'Job degree requirements',
    'job_health_insurance': 'Health insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(12, 5)

for i, (column, title) in enumerate(dict_column.items()):
    value_counts = df_ML_EGY[column].value_counts()
    labels = value_counts.index.astype(str)  # Get matching labels dynamically
    
    ax[i].pie(value_counts, labels=labels, autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.tight_layout()  # Prevent overlap
plt.show()


```

```python
df_plot = df_ML_EGY['company_name'].value_counts().head(10).to_frame()
#1. First thing we need to specify is DATA
#2. Second thing we need to specify is the X value,
# The x output will be shown horizontally/down so in this case we should put the count
#3. Third thing we need to specify is the Y value
# The y output will be shown vertically/side so in this case we should put the company_name
# 4. Fourth thing we need to specify is the hue value
# The hue value will be the color of the bar, in this case we will use the count as well
# 5. Fifth thing we need to specify is the palette
# The palette is the color of the hue, in this case we will use the dark:b_r
sns.set_theme(style='ticks')
sns.despine()
plt.title('Counts of companies of Machine Learning in Egypt')
plt.ylabel('')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
plt.show()
```
